```python
import random

correct_secret = 0
ones_or_thirty_six = 0
correct_relationship = 0
errors = 0
runs = 100000

for i in range(runs):
    #set up random variables
    d = 7
    q = random_prime(2000, lbound=100)
    Z37 = Integers(q)
    Z3 = Integers(3)
    a = []
    e = []
    
    for i in range(d):
        a.append(Z37.random_element())
        e.append(Integer((Z3.random_element()))-1)

    ZX37.<s> = PolynomialRing(Z37)

    secret = Z37.random_element()

    p = [Z37(0)] * d
    b = [Z37(0)] * d
    newc = []
    
    for i in range(len(a)):
        #could it be here that the construction fails??
        b[i] = (a[i] * secret) + e[i]
        p1 = ZX37(b[i] - (a[i] * s))
        p2 = ZX37(b[i] - (a[i] * s + 1))
        p3 = ZX37(b[i] - (a[i] * s - 1))
        p[i] = (p1 * p2 * p3)

        c = []
        for coeff in p[i]: 
            c.append(coeff)
        if(c != []):
            c.reverse()
            newc.append(c)
    rows = len(newc)
    
    M = MatrixSpace(Z37, rows, len(newc[0]))
    M2 = MatrixSpace(Z37, rows, 1)
    
    
    try:
        C = M(newc)
    except ValueError:
        errors = errors + 1
    
    cols = len(C[0])
    
    M3 = MatrixSpace(Z37, cols, 1)
    
    checker = []
    for i in range(cols):
        checker.append(secret**(i))
    checker = M3(checker.reverse())
    zeros = M2()
    
    if((C * checker) != zeros):
        print((C * checker))
    
    #v = M2([0] * rows)
    #v.transpose()
    #C = C.augment(v)

    C = C.echelon_form()
    

    u1 = q - C[2, 3]
    u2 = q - C[1, 3]
    u3 = q - C[0, 3]
    

    if u1 == secret:
        correct_secret = correct_secret + 1
    else:
        print(u1, "\t", secret)
    if (u2 == u1 ** 2) and (u3 == u1 ** 3):
        correct_relationship = correct_relationship + 1
        
print("Secret correctly guessed: \t\t\t",correct_secret * 1.0 / runs)
print("Linearized variables are correctly related: \t",correct_relationship * 1.0 / runs )
print("Number of errors: \t\t\t\t",errors)
```
